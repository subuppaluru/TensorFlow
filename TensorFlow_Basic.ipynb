## Tensors

```python
hello = tf.constant('Hello')
```

```python
print(hello)
print(type(hello))
```

```python
world = tf.constant('World')

print(world)
print(type(world))
```

```python
result = hello +" "+ world

print(result)
print(type(result))
```

** Computations **

```python
tensor_1 = tf.constant(1)
tensor_2 = tf.constant(2)

print(tensor_1)
print(type(tensor_1))
print(tensor_2)
print(type(tensor_2))
```

```python
print(tensor_1 + tensor_2)
```

## Operations

```python
const = tf.constant(10)

print(const)
```

```python
fill_mat = tf.fill((4,4),10)

print(fill_mat)

print(type(fill_mat))
```

```python
myzeros = tf.zeros((4,4))

print(myzeros)

print(type(myzeros))
```

```python
myones = tf.ones((4,4))

print(myones)

print(type(myones))
```

```python
my_ops = [const,fill_mat,myzeros,myones]

print(my_ops)

print(type(my_ops))
```

## Matrix Multiplication

```python
a = tf.constant([ [1,2],
                  [3,4] ])

print(a)

print(a.get_shape())
```

```python
b = tf.constant([[10],[100]])

print(b)
print(b.get_shape())
```

```python
result = tf.matmul(a,b)

print(result)

print(result.get_shape())
```
